```python
import pandas as pd

max_pow = 3
_b = 3
def f(x,p_max,b=_b,C=1):
    truth = []
    for d in range(p_max):
        factor = b**(d+1)
        factor_prime = b**d
        c = int((x %factor)/factor_prime)
        truth.append(c == C)
    return np.array(truth)

def f_row(row):
    return np.any(f(row["x"], max_pow) &  f(row["y"], max_pow))
    
L = _b**max_pow
coords = []
for i in range(L):
    for j in range(L):
        coords.append([i,j])
df = pd.DataFrame(coords,columns=["x","y"])
df["c"] = df.apply(f_row,axis=1)

df.plot(kind='scatter',x='x', y='y',  c='c', s=50,figsize=(15,10))
```

```python
import pandas as pd
import warnings
import numpy as np
def to_points(v_str,D=2,L=127):
    arr = v_str.split(' ')
    arr = np.array([int(s) for s in np.array(arr) if s != ""])
    ay= arr % L
    ax = (arr / L).astype(int)
    return ax,ay

def to_mask(v_str,D=2, L=127):
    image = np.ones((L,L))
    arr = v_str.split(' ')
    arr = np.array([int(s) for s in np.array(arr) if s != ""])
    ay= arr % L
    ax = (arr / L).astype(int)
    image[ax,ay] = 0
    return image
    
%matplotlib inline
warnings.filterwarnings('ignore')
df = pd.read_csv("../BWS_VS_PROJECT/BWS/out.data", sep="~",names = ["DATA"], header=None)
df.head()
```

```python
def cut(line, distance):
    # Cuts a line in two at a distance from its starting point
    if distance <= 0.0 or distance >= line.length:
        return [LineString(line)]
    coords = list(line.coords)
    for i, p in enumerate(coords):
        pd = line.project(Point(p))
        if pd == distance:
            return [
                LineString(coords[:i+1]),
                LineString(coords[i:])]
        if pd > distance:
            cp = line.interpolate(distance)
            return [
                LineString(coords[:i] + [(cp.x, cp.y)]),
                LineString([(cp.x, cp.y)] + coords[i:])]
```

```python
def lagged_jump(ref_row):
    """assumes a spatial matrix - all cols are coords
       computes the euclidean distance between two rows    
    """
    
    def _lagged_jump(row): return euclidean(list(ref_row),list(row))
    return _lagged_jump
    
def row_fix( a):
    #assumed square distance matrix so lengths are predictable
    """
    some sort of padding - i cant recall what i was doing... look into it
    """
    a = a[np.isfinite(a)]
    temp = np.zeros(len(a))
    temp[:len(a)] = a
    return a

def shift_matrix(mat,stack_it=True):
    """
    a sort of pivot where we shigt the lag matrix to correct triangular form so that lags are in columns
    """
    new_rows = []
    for i in range(mat.shape[0]):   new_rows.append(row_fix(mat[:,i]))

    sds = pd.DataFrame(new_rows)
    sds = np.abs(sds)
    sds.index.name='l'
    
    if stack_it == True:
        stacked = sds.stack().reset_index()
        stacked.columns = ["time", "lag", "sdisp"]                
        return stacked
    
    return sds

def track_displacements(single_track,stack=True):
    """
    Given a single track, generate the lagged displacement matrix
    """
    result = pd.DataFrame(index=single_track.index)
    for i in range(len(result)):  result[str(i)] = single_track.apply(lagged_jump(single_track.iloc[i]),axis=1)

    #get diff matrix
    upt = np.tril(np.ones(result.shape)).astype(np.bool)
    results_flat = result.where(upt)
    results_flat.index.name = "_"
    mat = results_flat.as_matrix()
    
    #create the shifted, stacked matrix
    stacked = shift_matrix(mat,stack)
    #t_index = stacked["lag"].unique()
    #t_index.sort()
    return stacked
```
